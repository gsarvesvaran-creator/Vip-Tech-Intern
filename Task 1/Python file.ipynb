```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
cgpa = float(input("Enter your CGPA: "))
placed = input("Placed? (True/False): ") == "True"

print("\n===== STUDENT PROFILE =====")
print(f"Name      : {name}")
print(f"Age       : {age}")
print(f"CGPA      : {cgpa}")
print(f"Placed    : {placed}")

print("\n===== DATA TYPES =====")
print(type(name))
print(type(age))
print(type(cgpa))
print(type(placed))
```

```python
x = int(input("Enter the 1st number : "))
y = int(input("Enter the 2nd number : "))

print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)

if y!= 0:
    print("Division:", x / y)
    print("Floor Division:", x // y)
    print("Modulo:", x % y)
else:
    print("Division, Floor Division, and Modulo cannot be performed with 0.")

print("Power:", x ** y)
```

```python
df = df.drop(columns=[
    "thumbnail",
    "Unnamed: 0",
    "work_from_home",
    "salary",
    "salary_pay",
    "salary_rate",
    "salary_avg",
    "salary_min",
    "salary_max",
    "salary_hourly",
    "salary_yearly",
    "salary_standardized",
    "commute_time"
])

df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
print(df["location"].value_counts().head(20))
```

```python
df["location"] = (
    df["location"]
    .astype(str)
    .str.replace("\xa0", " ", regex=False)   # Remove non-breaking spaces
    .str.replace(r"\s+", " ", regex=True)    # Replace multiple spaces with one
    .str.strip()                             # Remove leading/trailing spaces
    .str.title()                             # Standardize case
)
```

```python
print(df["location"].value_counts().head(20))
```

```python
# Count Jobs by Company
company_jobs = (
    df.groupby("company_name")["job_id"]
      .count()
      .reset_index(name="Total_Jobs")
      .sort_values(by="Total_Jobs", ascending=False)
)

company_jobs.head(10)
```

```python
# Count Jobs by Location
location_jobs = (
    df.groupby("location")["job_id"]
      .count()
      .reset_index(name="Total_Jobs")
      .sort_values(by="Total_Jobs", ascending=False)
)

location_jobs.head(10)
```

```python
# Company-wise Multiple Statistics
company_summary = (
    df.groupby("company_name")
      .agg(
          Total_Jobs=("job_id", "count"),
          Unique_Locations=("location", "nunique"),
          Job_Titles=("title", "nunique")
      )
      .sort_values(by="Total_Jobs", ascending=False)
)

company_summary.head(10)
```

```python
# Add Company Total Jobs to Every Row
df["Company_Total_Jobs"] = (
    df.groupby("company_name")["job_id"]
      .transform("count")
)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
plt.plot(company_jobs.index, company_jobs.values, marker='o')
plt.title("Top 10 Companies by Job Openings")
plt.xlabel("Company Name")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(location_jobs.index, location_jobs.values)
plt.title("Top 10 Job Locations")
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(company_summary["Total_Jobs"],
            company_summary["Unique_Titles"])

plt.title("Company Jobs vs Unique Job Titles")
plt.xlabel("Total Jobs")
plt.ylabel("Unique Job Titles")
plt.show()
```

```python
top_company = (
    df.groupby("company_name")["job_id"]
      .count()
      .sort_values(ascending=False)
      .head(5)
)

plt.figure(figsize=(7,7))
plt.pie(top_company.values,
        labels=top_company.index,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Top 5 Hiring Companies")
plt.show()
```

```python
description_length = df["description"].str.len()

plt.figure(figsize=(8,5))
plt.hist(description_length, bins=30)
plt.title("Distribution of Job Description Length")
plt.xlabel("Characters")
plt.ylabel("Frequency")
plt.show()
```

```python
company_jobs = (
    df.groupby("company_name")["job_id"]
      .count()
      .sort_values(ascending=False)
      .head(10)
      .reset_index(name="Total_Jobs")
)

plt.figure(figsize=(10,5))
sns.barplot(data=company_jobs,
            x="company_name",
            y="Total_Jobs")

plt.title("Top 10 Hiring Companies")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Description_Length"] = df["description"].str.len()

top_location = df["location"].value_counts().head(10).index

plt.figure(figsize=(12,6))

sns.boxplot(
    data=df[df["location"].isin(top_location)],
    x="location",
    y="Description_Length",
    showfliers=False
)

plt.xticks(rotation=45)
plt.title("Description Length by Top Locations")
plt.xlabel("Location")
plt.ylabel("Description Length")
plt.show()
```

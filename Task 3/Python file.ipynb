```python
df = df.drop(columns=[
    "thumbnail",
    "Unnamed: 0",
    "work_from_home",
    "salary",
    "salary_pay",
    "salary_rate",
    "salary_avg",
    "salary_min",
    "salary_max",
    "salary_hourly",
    "salary_yearly",
    "salary_standardized",
    "commute_time"
])

df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
print(df["location"].value_counts().head(20))
```

```python
df["location"] = (
    df["location"]
    .astype(str)
    .str.replace("\xa0", " ", regex=False)   # Remove non-breaking spaces
    .str.replace(r"\s+", " ", regex=True)    # Replace multiple spaces with one
    .str.strip()                             # Remove leading/trailing spaces
    .str.title()                             # Standardize case
)
```

```python
print(df["location"].value_counts().head(20))
```

```python
# Count Jobs by Company
company_jobs = (
    df.groupby("company_name")["job_id"]
      .count()
      .reset_index(name="Total_Jobs")
      .sort_values(by="Total_Jobs", ascending=False)
)

company_jobs.head(10)
```

```python
# Count Jobs by Location
location_jobs = (
    df.groupby("location")["job_id"]
      .count()
      .reset_index(name="Total_Jobs")
      .sort_values(by="Total_Jobs", ascending=False)
)

location_jobs.head(10)
```

```python
# Company-wise Multiple Statistics
company_summary = (
    df.groupby("company_name")
      .agg(
          Total_Jobs=("job_id", "count"),
          Unique_Locations=("location", "nunique"),
          Job_Titles=("title", "nunique")
      )
      .sort_values(by="Total_Jobs", ascending=False)
)

company_summary.head(10)
```

```python
# Add Company Total Jobs to Every Row
df["Company_Total_Jobs"] = (
    df.groupby("company_name")["job_id"]
      .transform("count")
)

df.head()
```
